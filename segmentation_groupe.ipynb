```python
import pandas as pd
```

```python
df = pd.read_excel("GRC_Spotify_Exo_1.xlsx")
#nombre de samples in the dataset 
n = len(df)
```

```python
#extraire les email 
prenoms_feminins = ["lea", "alice", "claire", "emma"]
prenoms_masculins = ["jean", "martin", "lucas", "olivier"]

#fonction pour detecter le sexe
def detection_sexe(email):
    prenom = email.split('.')[0].lower()
    if prenom in prenoms_feminins:
        return "Féminin"
    elif prenom in prenoms_masculins:
        return "Masculin"
    else:
        return "Inconnu"

df["sexe"] = df["Email"].apply(detection_sexe)

```

```python
segment_group_Age = df["Age"].value_counts()
segment_group_Profession = df["Profession"].value_counts()
segment_groupby_Region = df["Region"].value_counts()
segment_groupby_sex = df["sexe"].value_counts()
```

```python
#intervale d'age 

age_18_24 = len(df[(df["Age"]>=18) & (df["Age"] < 25)])
age_25_30 = len(df[(df["Age"]>=25) & (df["Age"] < 31)])
age_sup_30 = len(df[(df["Age"]>=25) & (df["Age"] >= 31)]) 
print("age entre 18 et 24 and",age_18_24)
print("age entre 25 et 30", age_25_30)
print(print("Age sup a 30 ans", age_sup_30))
print("segment_group_Profession :  ", segment_group_Profession)
print("segment_groupby_Region : ", segment_groupby_Region)
print("segment_groupby_sex : ",segment_groupby_sex)
```
